```python
import numpy as np
import pandas as pd
```

```python
def extract_data(path):
  return pd.read_csv(path)
```

```python
def transform_data(df):
  #So original dataset doesn’t get modified
  df=df.copy()
  #inplace=True same dataframe m change that is we created copy of df
  df["Age"].fillna(df["Age"].mean(),inplace=True)
  df["Embarked"].fillna(df["Embarked"].mode()[0],inplace=True)
  #too many missing values
  df.drop(columns=["Cabin"],inplace=True)
  df.drop_duplicates(inplace=True)
  df["Sex_Encoded"]=df["Sex"].map({"male":0,"female":1})
  #👉 One-Hot Encoding
  #It converts one column into multiple binary (0/1) columns.
  #Dummies = new columns made of 0 and 1 from text categories
  df=pd.get_dummies(df,columns=["Embarked"],drop_first=True)
  df["Family_Size"]=df["SibSp"]+df["Parch"]+1
  return df
```

```python
df_cleaned=transform_data(extract_data("titanic.csv"))
```

```python
df_cleaned.to_csv("titanic_cleaned.csv",index=False);
```
